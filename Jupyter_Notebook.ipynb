```python
# Importing the required packages
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Firstly, Preparing training data

train_df = pd.read_csv("https://raw.githubusercontent.com/amirbek-akramov/AviaTicketsPredictorProject/main/train_data.csv")
train_df
```

```python
# Shape of the training dataframe
train_df.shape
```

```python
# Info about training dataframe
train_df.info()
```

```python
# Describtion on the training dataframe
train_df.describe()
```

```python
train_df.isnull().sum()
```

```python
train_df.fillna(method='bfill')
```

```python
str_columns =  ["airline", 'flight', 'source_city', 'departure_time', 'stops', 'arrival_time', 'destination_city', 'class']

# LabelEncoder
label_encoder = LabelEncoder()


for column in str_columns:
    train_df[column] = label_encoder.fit_transform(train_df[column])

train_df
```

```python
# Preparing the test data
test_df = pd.read_csv("https://raw.githubusercontent.com/amirbek-akramov/AviaTicketsPredictorProject/main/test_data.csv")
test_df
```

```python
# Shape of the test dataframe
test_df.shape
```

```python
# Info about test dataframe
test_df.info()
```

```python
# Describtion on the test dataframe
test_df.describe()
```

```python
test_df.isnull().sum()
```

```python
# LabelEncoder
label_encoder = LabelEncoder()


for column in str_columns:
    test_df[column] = label_encoder.fit_transform(test_df[column])

test_df
```

```python
train_df.drop(columns='id', inplace=True)
```

```python
# Splitting the data

# !!! If you have problems with RAM in google colab, use this code
# train_df = train_df.iloc[:7000]

X = train_df.drop(columns='price')
y = train_df['price']
# I want to use both test and train data but it will be more effective if I use only train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Training model
RF = RandomForestClassifier()
```

```python
# Fit the RandomForest
RF.fit(X_train, y_train)
```

```python
y_pred = RF.predict(X_test)
```

```python
# Evaluate the RandomForest model
print("\nModel Evaluation:")
print("Accuracy Score:", round(metrics.accuracy_score(y_test, y_pred)*100), "%")
print("RMSE:", round(metrics.mean_squared_error(y_test, y_pred, squared=False)))
print("MAE:", round(metrics.mean_absolute_error(y_test, y_pred)))
```

```python
conf_mat = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()
```

```python
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```
